```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
documents = ["people watch movies",
             "people like movies",
             "people like music",
             "people watch music"]
```

```python
tokens = [sentence.lower().split() for sentence in documents]
print(tokens)
```

```python
all_words = [word for sentence in tokens for word in sentence]
print(all_words)
unique_words = set(all_words)
print(unique_words)
```

```python
encoder = OneHotEncoder(sparse_output=False)
# encoder.fit(all_words)
```

```python
encoder.fit(np.array(all_words).reshape(-1, 1))
```

```python
print("vocabulary:", encoder.categories_[0])
```

```python


for sentence in tokens:
    encoded_sentence = encoder.transform(np.array(sentence).reshape(-1, 1))
    print(sentence)
    print(encoded_sentence)
```

```python
#Bag of words

#Bag or words captures how frequently a word us used in a meaningful way. 


# Based on the count of words. word occurence in a document is more then it is more important.

# some of the examples of the bag of words are below: 

#vocabulary: ['like' 'movies' 'music' 'people' 'watch']. ==> unique words in the document.

#if we preprocess we will remove words like 'a' , 'the' , 'is' etc. these are called stop words.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movies and music",
             "people like movies and like music",
             "people like music and watch music and watch cricket",
             "people watch music like cricket watch movies"]
```

```python
bow = CountVectorizer()
bow_transformed = bow.fit_transform(documents)
print("vocabulary:", bow.get_feature_names_out())
print(bow_transformed.toarray())
```

```python
vocab = bow.get_feature_names_out()
matrix = bow_transformed.toarray()

print("Vocabulary:", vocab)
print("\nBag-of-words matrix (rows=document, columns=word counts):")
print(matrix)

for i, doc in enumerate(documents):
    print(f"\nDocument {i+1}: {doc}")
    for word, count in zip(vocab, matrix[i]):
        if count > 0:
            print(f"  {word}: {count}")
```

#TF-IDF

#(Term Frequency - Inverse Document Frequency) is a statistical measure used to evaluate the importance of a word 
#in a document relative to a collection of documents (corpus). 
#It helps to identify words that are more relevant to a specific document while downplaying common words that appear frequently across all documents.

```python
TF(Word, D) = (Occurrences of the word in the sentencec) / (total words in the given sentence)

IDF(Word) = log(Total number of sentence / Number of sentence containing the word)
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

new_documents = [
    "machine learning helps computers learn from data",
    "deep learning is a subset of machine learning",
    "natural language processing works with text DATA",
    "data science uses statistics and machine learning"
]

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(new_documents)

print("Vocabulary:", tfidf.get_feature_names_out())
print("\nTF-IDF matrix:")
print(tfidf_matrix.toarray())

for i, doc in enumerate(new_documents):
    print(f"\nDocument {i+1}: {doc}")
    for word, score in zip(tfidf.get_feature_names_out(), tfidf_matrix.toarray()[i]):
        if score > 0:
            print(f"  {word}: {score:.3f}")
```
